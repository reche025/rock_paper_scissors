```python
print("\nPart 1:\n")

import random
player_name = input("Welcome!\n\nWhat is your name?\n") 
options = ["stone", "paper", "scissors"]
r = 1
x = 0
y = 0

game_count = input("\nHow many matches would you like to play?\n")

def game_engine():
  global x
  global y
  choice = input("\nWhat is your choice: (stone, paper or scissors)\n")
  while choice.lower() != "stone" and choice.lower() != "paper" and choice.lower() != "scissors":
    choice = input("\nSeems like your response did not match one of the options, please enter a correct choice: (stone, paper or scissors)\n")
  machines_choice = random.choice(options)
  print("\nThe machine chooses:", machines_choice)
  if machines_choice == choice:
    print("\nIt's a draw!")
  else:
    if machines_choice == "stone" and choice == "paper":
      print("It's your point!")
      x += 1 
    elif machines_choice == "paper" and choice == "scissors":
      print("It's your point!")
      x += 1
    elif machines_choice == "scissors" and choice == "stone":
      print("It's your point!")
      x += 1
    elif choice == "stone" and machines_choice == "paper":
      print("It's the machine's point!")
      y += 1
    elif choice == "paper" and machines_choice == "scissors":
      print("It's the machine's point!")
      y += 1
    elif choice == "scissors" and machines_choice == "stone":
      print("It's the machine's point!")
      y += 1
  print("\n" + player_name + " has", x, "points")
  print("The Machine has", y, "points")

if int(game_count) == 1:
  print("\nYou only have 1 chance to defeat the machine!")
elif int(game_count)%2 == 0:
  print("\nYou must win", int((int(game_count)/ 2) + 1), "matches to defeat the machine") 
elif int(game_count)%2 > 0:
  print("\nYou must win", int((int(game_count)/ 2) + 0.5), "matches to defeat the machine") 

if int(game_count) == 1:
  while x < 1 or y < 1:
    game_engine()
    if x == 1:
      print("\n" + player_name + " is the winner!")
      break
    elif y == 1:
      print("\nThe machine defeated you!")
      break
elif int(game_count)%2 == 0:
  while x < int((int(game_count)/ 2) + 1) or y < int((int(game_count)/ 2) + 1):
    game_engine()
    if x == int((int(game_count)/ 2) + 1):
      print("\n" + player_name + " is the winner!")
      break
    elif y == int((int(game_count)/ 2) + 1):
      print("\nThe machine defeated you!")
      break
elif int(game_count)%2 > 0:
  while x < int((int(game_count)/ 2) + 0.5) or y < int((int(game_count)/ 2) + 0.5):
    game_engine()
    if x == int((int(game_count)/ 2) + 0.5):
      print("\n" + player_name + " is the winner!")
      break
    elif y == int((int(game_count)/ 2) + 0.5):
      print("\nThe machine defeated you!")
      break
```
